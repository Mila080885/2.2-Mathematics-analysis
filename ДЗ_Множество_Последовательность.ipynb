## Тема “Множество”
1. Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.
Выполнить задание 1 на языке Python

```python
a = {1,2,3}
b = {2,3,4}
c = {1,4,8}
```

Пересечение множеств (члены обоих множеств)

```python
a & b
a.intersection(b)
```

```python
a & c
a.intersection(c)
```

```python
c & b
c.intersection(b)
```

Объединение (члены обоих множеств)

```python
a | b
a.union(b)
```

```python
a | c
a.union(c)
```

```python
c | b
c.union(b)
```

Разность множеств (члены только первого множества, но не второго)

```python
a - b
a.difference(b)
```

```python
b - a
b.difference(a)
```

```python
a - c
a.difference(c)
```

```python
c - a
c.difference(a)
```

```python
c - b
c.difference(b)
```

```python
b - c
b.difference(c)
```

Исключающее ИЛИ (элементы или первого, или второго множества, но не общие)

```python
a ^ b
a.symmetric_difference(b)
```

```python
a ^ c
a.symmetric_difference(c)
```

```python
b ^ c
b.symmetric_difference(c)
```
